# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self,marka,hiz):
        self.marka = marka
        self.hiz = hiz
class Bus(Vehicle):
    def __init__(self, marka, hiz,seating_capacity=50):
      super().__init__(marka, hiz) 
      self.seating_capacity = seating_capacity
    
bus1 = Bus("Mercedes",200)
print(f"Brand of a bus: {bus1.marka}")  
print(f"max velocity of a bus: {bus1.hiz} km/saat")
print(f"The seating capacity is {bus1.seating_capacity} passengers.") 
````

School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    def __init__(self,marka,hiz):
        self.marka = marka
        self.hiz = hiz
class schoolBus(Vehicle):
    def __init__(self, marka, hiz,renk,yas):
        super().__init__(marka, hiz)
        self.renk= renk
        self.yas = yas 
schoolBus1 = schoolBus("MAN",200,"Beyaz",12)

def isHeritance():
    if isinstance(schoolBus1, Vehicle):
        print(f"""
    {schoolBus1.marka} marka okul otobüsü maksimum {schoolBus1.hiz} km/saat hız yapar.
    {schoolBus1.renk} renginde ve {schoolBus1.yas} yaşındadır
    Vehicle sınfının bir örneğidir.""")
    else:
        print("araç vehicle sınfının bir örneği değildir.")
isHeritance()
````

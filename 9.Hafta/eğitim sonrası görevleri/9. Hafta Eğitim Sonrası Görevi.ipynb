Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
tvKanalları = {
"Kanal D" : 1,
"ATV" : 2,
"NOW" : 3,
"SHOW TV" : 4

}

class kumanda():

    def __init__(self,tv_durum = "kapalı", ses = 0,kanal_listesi = ["Kanal D", "ATV", "NOW", "SHOW TV"], kanal_secim = 0):
        self.tv_durum = tv_durum
        self.ses = ses
        self.kanal_listesi = kanal_listesi
        self.kanal_secim = kanal_secim
        self.tvKanalları = tvKanalları


    def tvAc(self):
        import time
        if self.tv_durum == "kapalı":
            print("""
                  TV açılıyor..
                  
                  """)
            time.sleep(1)
            self.tv_durum = "açık"
        else:
            print("""
                TV zaten açık  
                  
                    """)
            

    def tvKapa(self):
        import time 
        if self.tv_durum == "açık":
            print("""
                 TV kapanıyor... 
                  """)
            time.sleep(1)
            self.tv_durum = "kapalı"
        else:
            print("""
                 TV zaten kapalı 
                  """)
            
            
    def sesDegistirme(self):
        import time
        print(f"ses düzeyi = {self.ses}")
        sesAcma = input("sesi düzeyini arttırmak istiyorsanız 'arttır' azaltmak istiyorsanız 'azalt' yazınız.")
        if self.tv_durum == "kapalı":
            print("önce televizyonu açınız")
        elif sesAcma.lower() == "arttır":
            sesEkle = int(input("sesi ne kadar arttırmak istiyorsunuz?"))
            self.ses += sesEkle
            time.sleep(1)
            print(f"ses düzeyi : {self.ses}") 
        elif sesAcma.lower() == "azalt":
            sesEkleme = int(input("ses düzeyini ne kadar azaltmak istiyorsunuz?"))
            self.ses -= sesEkleme
            time.sleep(1)
            print(f"ses düzeyi : {self.ses}")
        else:
            print("lütfen azalt ya da arttır yazınız")

    def kanalEkle(self):
        if self.tv_durum == "kapalı":
            print("""
                  önce TV'yi açınız.
                  """)
        else:
            for item in list(tvKanalları.keys()):
                print(f"{item.upper()} : {tvKanalları[item]} ")
            yeni_kanal = input("eklemek istediğiniz kanalın adını giriniz:")
            yeni_konum = input("eklediğiniz kanal kaçıncı konumda?")
            tvKanalları[yeni_kanal] = [yeni_konum]

    def kanal_listesi(self):
        if self.tv_durum == "kapalı":
            print("""
                  Önce TV'yi açınız.
                 """)
        else:
            for i in list(tvKanalları.keys()):
                print(f"{i.upper()} : {tvKanalları[i]}")

    def rastgeleKanal(self):                                     #BU FONKSİYON DÜZENLENMELİ!!!
        import random
        if self.tv_durum == "kapalı":
            print("önce TV'yi açınız")
                
        rastgele = input("rastgele bir kanala geçmek için R'ye basınız")

        elif rastgele == "R":
            for i in random.randint(1,4):
                if i == 1:                                    
                    print("Şu anda Kanal D'desiniz.")   
                elif i == 2:
                    print("Şu anda ATV'desiniz.")
                elif i == 3:
                    print("Şu anda NOW'dasınız.")
                else:
                    print("Şu anda SHOW TV'desiniz.")
    

kumanda1 = kumanda()

print("""
      
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkmak için q'ya basınız.
    """)

while True:
    islem = (input("lütfen işlemi seçiniz."))
    if islem == "q":
        print("sistem kapanıyor...")
        break
    elif islem == "1":
        kumanda1.tvAc()
    elif islem == "2":
        kumanda1.tvKapa()
    elif islem == "3":
        kumanda1.sesDegistirme()
    elif islem == "4":
        kumanda1.kanalEkle()
    elif islem == "5":
        pass
    elif islem == "6":
        pass
    elif islem == "7":
        kumanda1.kanal_listesi()
    else:
        print("lütfen geçerli bir işlem giriniz.")
```

1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class kumanda():

    def __init__(self,tv_durum = "kapalı", ses = 0,kanal_listesi = ["Kanal D", "ATV", "NOW", "SHOW TV"], kanal_secim = "SHOW TV"):
        self.tv_durum = tv_durum
        self.ses = ses
        self.kanal_listesi = kanal_listesi
        self.kanal_secim = kanal_secim
        


    def tvAc(self):
        if self.tv_durum == "kapalı":
            print("""
                  TV açılıyor..
                  
                  """)
            self.tv_durum = "açık"
        else:
            print("""
                TV zaten açık  
                  
                    """)
            

    def tvKapa(self):
        if self.tv_durum == "açık":
            print("""
                 TV kapanıyor... 
                  """)
            self.tv_durum = "kapalı"
        else:
            print("""
                 TV zaten kapalı 
                  """)
            
            
    def sesDegistirme(self):
        
        print(f"ses düzeyi = {self.ses}")
        sesAcma = input("sesi düzeyini arttırmak istiyorsanız 'arttır' azaltmak istiyorsanız 'azalt' yazınız.")
        if self.tv_durum == "kapalı":
            print("önce televizyonu açınız")
        elif sesAcma.lower() == "arttır":
            sesEkle = int(input("sesi ne kadar arttırmak istiyorsunuz?"))
            self.ses += sesEkle
            print(f"ses düzeyi : {self.ses}") 
        elif sesAcma.lower() == "azalt":
            if self.ses != 0 :
                sesEkleme = int(input("ses düzeyini ne kadar azaltmak istiyorsunuz?(0'ın altınaa düşmeyiniz)"))
                self.ses -= sesEkleme
                print(f"ses düzeyi : {self.ses}")
        else:
            print("lütfen azalt ya da arttır yazınız")

    def kanalEkle(self, kanal_ismi):
        print("kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal_ismi)
        print("kanal eklendi.")

    def rastgeleKanal(self):                                     #BU FONKSİYON DÜZENLENMELİ!!!
        import random
        if self.tv_durum == "kapalı":
            print("önce TV'yi açınız")
        else:      
            rastgele = random.randint(0,len(self.kanal_listesi)-1)
            self.kanal_secim = self.kanal_listesi[rastgele]
            print("şu anki konum" , self.kanal_secim)
        
    def __len__(self):
        return len(self.kanal_listesi)
    
    def __str__(self):
        return(f"\n Tv durum: {self.tv_durum}\n Tv ses: {self.ses}\n Kanal listesi : {self.kanal_listesi}\n Şu an bu kanaldasınız : {self.kanal_secim}")

   
    

kumanda1 = kumanda()

print("""
      
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkmak için q'ya basınız.
    """)

while True:
    islem = (input("lütfen işlemi seçiniz."))
    if islem == "q":
        print("sistem kapanıyor...")
        break
    elif islem == "1":
        kumanda1.tvAc()
    elif islem == "2":
        kumanda1.tvKapa()
    elif islem == "3":
        kumanda1.sesDegistirme()
    elif islem == "4":
        kanal_isimleri = input("kanal isimlerini ', ' ile ayırarak giriniz.")
        kanal_listesi = kanal_isimleri.split(",")
        for i in kanal_listesi:
            kumanda1.kanalEkle(i)        
    elif islem == "5":
        print("kanal sayısı:" , len(kumanda1))

    elif islem == "6":
        kumanda1.rastgeleKanal()
    elif islem == "7":
        print(kumanda1)
    else:
        print("lütfen geçerli bir işlem giriniz.")
```

```python
import random
import time            

class kumanda():
    def __init__(self,tv_durum = "kapalı",tv_ses = 0, kanal_listesi = ["Trt"] , kanal = "Trt"):
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi
        self.kanal = kanal

    def tv_ac(self):
        if (self.tv_durum == "açık"):
            print("tv zaten açık")
        else:
            print("tv açılıyor...")
            self.tv_durum = "açık"
    
    def tv_kapat(self):
        if (self.tv_durum == "kapalı"):
            print("tv zaten kapalı")
        else:
            print("tv kapanıyor...")
            self.tv_durum = "kapalı"

    def ses_ayarları(self):
        while True:
            cevap = input("sesi arttırmak için > yaz, azaltmak için < yaz, çıkış yapmak için çıkış yaz.")
            if cevap == "<":
                if self.tv_ses != 0:
                    self.tv_ses -= 1
                    print("ses:", self.tv_ses)
            elif cevap == ">":
                if self.tv_ses != 20:
                    self.tv_ses += 1
                    print("ses:", self.tv_ses)
            else:
                print("ses güncellendi.ses:", self.tv_ses)
                break
    def kanal_ekle(self,kanal_ismi):
        print("kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal_ismi)
        print("kanal eklendi...")

    def rastgele_kanal(self):
        rastgele = random.randint(0,len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele]    
        print("şu anki kanal:" , self.kanal)

    def __len__(self):
        return len(self.kanal_listesi)
    
    
    def __str__(self):
        return f"Tv durum:{self.tv_durum}\n Tv ses: {self.tv_ses} \n Kanal Listesi : {self.kanal_listesi} \n Şu an bulunulan kanal: {self.kanal}"
    
kumanda = kumanda()

print("""
      
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkmak için q'ya basınız.
    """)

while True:
    islem = input("Lütfen yapmak istediğiniz işlemi giriniz.")
    if islem == "q":
        print(("program sonlandırılıyor..."))
        break
    elif islem == "1":
        kumanda.tv_ac()

    elif islem == "2":
        kumanda.tv_kapat()

    elif islem == "3":
        kumanda.ses_ayarları()
    
    elif islem == "4":
        kanal_isimleri = input("kanal isimlerini , ile ayırarak girin")
        kanal_listesi = kanal_isimleri.split(",")
        for i in kanal_listesi:
            kumanda.kanal_ekle(i)

    elif islem == "5":
        print("kanal sayısı:" , len(kumanda))

    elif islem  == "6":
        kumanda.rastgele_kanal()

    elif islem == "7":
        print(kumanda)

    else:
        print("geçersiz işlem!")
```

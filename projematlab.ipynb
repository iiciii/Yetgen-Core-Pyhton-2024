```python
[~, ~, coffeeData] = xlsread('Fortunes.xlsx', 'CoffeeFortuneFigures');
[~, ~, coffeeRandom] = xlsread('Fortunes.xlsx', 'CoffeeFortuneRandom');
[~, ~, cardsData] = xlsread('Fortunes.xlsx', 'CardsFortune');

% Welcome message
disp('Welcome to Falci Baci!')

% Prompt the user for the type of fortune
fortuneType = input('Please enter the type of fortune (1:Coffee fortune, 2:Cards fortune): ');

if fortuneType == 1
    % Coffee fortune
    
    disp('You requested a coffee fortune.')
    disp('Please tell me the three things you see in your cup among this list:')
    disp('Horse, Horseshoe, Axe, Drum, Giraffe, Bird, Road, Parrot, Sward, Unicorn, Tear,')
    disp('Evileye, Key, Car, Fullheart, Volcano, Duck, Chair, Brokenheart, Rooster, Tree,')
    disp('Frog, Snake, Devil, Fish, Moon, Table, Baby, Aircraft, Kite:')
    
    % Prompt the user for three figures
    figure1 = input('Figure 1: ', 's');
    figure2 = input('Figure 2: ', 's');
    figure3 = input('Figure 3: ', 's');
    
    % Check if the figures are valid
    validFigures = intersect({figure1, figure2, figure3}, coffeeData(:,1));
    
    if numel(validFigures) < 3
        disp('One or more figures entered is invalid.')
    else
        % Find corresponding statements for figures
        idx1 = find(strcmp(coffeeData(:,1), figure1));
        idx2 = find(strcmp(coffeeData(:,1), figure2));
        idx3 = find(strcmp(coffeeData(:,1), figure3));

        % Randomly select two sentences
        randomIdx = randperm(size(coffeeRandom,1), 2);

        % Display the fortune
        disp(['>>' coffeeData{idx1, 2}])
        disp(coffeeRandom{randomIdx(1), 1})
        disp(['>>' coffeeData{idx2, 2}])
        disp(['>>' coffeeData{idx3, 2}])
    end
    
elseif fortuneType == 2
    % Card fortune
    disp('You requested a card fortune.')
    disp('Please tell me the type of cards fortune (1:Draw one card, 2:Draw three cards, 3:Draw six cards): ')
    
    cardType = input('Type: ');
    
    % Shuffle the deck
    deck = randperm(52);
    
    if cardType == 1
        % Draw one card
        disp('Shuffling and drawing a card for you.')
        card = deck(end);
        disp(['>>' cardsData{card, 1} ': ' cardsData{card, 2}])
    elseif cardType == 2
        % Draw three cards
        disp('Shuffling and drawing three cards for you.')
        for i = 1:3
            card = deck(end - i + 1);
            switch i
                case 1
                    disp(['>>Past: ' cardsData{card, 1} ': ' cardsData{card, 2}])
                case 2
                    disp(['>>Present: ' cardsData{card, 1} ': ' cardsData{card, 2}])
                case 3
                    disp(['>>Future: ' cardsData{card, 1} ': ' cardsData{card, 2}])
            end
        end
    elseif cardType == 3
        % Draw six cards
        disp('Shuffling and drawing six cards for you.')
        categories = {'Yourself', 'Your family', 'Your friends', 'What you expect', 'What you do not expect', 'The outcome'};
        for i = 1:numel(categories)
            card = deck(end - i + 1);
            disp(['>>' categories{i} ': ' cardsData{card, 1} ': ' cardsData{card, 2}])
        end
    else
        disp('Invalid card type.')
    end
else
    disp('Invalid choice.')
end
```

```python
import pandas as pd
import random

def coffee_fortune():
    coffee_data = pd.read_excel('Fortunes.xlsx', sheet_name='CoffeeFortuneFigures')
    coffee_random = pd.read_excel('Fortunes.xlsx', sheet_name='CoffeeFortuneRandom')

    print("You requested a coffee fortune.")
    print("Please tell me the three things you see in your cup among this list:")
    print("Horse, Horseshoe, Axe, Drum, Giraffe, Bird, Road, Parrot, Sward, Unicorn, Tear,")
    print("Evileye, Key, Car, Fullheart, Volcano, Duck, Chair, Brokenheart, Rooster, Tree,")
    print("Frog, Snake, Devil, Fish, Moon, Table, Baby, Aircraft, Kite:")

    figure1 = input("Figure 1: ")
    figure2 = input("Figure 2: ")
    figure3 = input("Figure 3: ")

    valid_figures = coffee_data[coffee_data['Figure'].isin([figure1, figure2, figure3])]

    if len(valid_figures) < 3:
        print("One or more figures entered is invalid.")
    else:
        for figure in [figure1, figure2, figure3]:
            statement = coffee_data[coffee_data['Figure'] == figure]['Statement'].values[0]
            print(f">> {statement}")

        random_sentences = random.sample(coffee_random['Sentence'].tolist(), 2)
        for sentence in random_sentences:
            print(sentence)

def card_fortune():
    cards_data = pd.read_excel('Fortunes.xlsx', sheet_name='CardsFortune')

    print("You requested a card fortune.")
    card_type = int(input("Please tell me the type of cards fortune (1:Draw one card, 2:Draw three cards, 3:Draw six cards): "))

    deck = list(range(1, 53))
    random.shuffle(deck)

    if card_type == 1:
        card = deck.pop()
        print(f">> {cards_data['Card'][card-1]}: {cards_data['Meaning'][card-1]}")
    elif card_type == 2:
        print("Shuffling and drawing three cards for you.")
        for i in range(3):
            card = deck.pop()
            if i == 0:
                print(f">> Past: {cards_data['Card'][card-1]}: {cards_data['Meaning'][card-1]}")
            elif i == 1:
                print(f">> Present: {cards_data['Card'][card-1]}: {cards_data['Meaning'][card-1]}")
            else:
                print(f">> Future: {cards_data['Card'][card-1]}: {cards_data['Meaning'][card-1]}")
    elif card_type == 3:
        print("Shuffling and drawing six cards for you.")
        categories = ['Yourself', 'Your family', 'Your friends', 'What you expect', 'What you do not expect', 'The outcome']
        for i, category in enumerate(categories):
            card = deck.pop()
            print(f">> {category}: {cards_data['Card'][card-1]}: {cards_data['Meaning'][card-1]}")
    else:
        print("Invalid card type.")

print("Welcome to Falci Baci!")
fortune_type = int(input("Please enter the type of fortune (1:Coffee fortune, 2:Cards fortune): "))

if fortune_type == 1:
    coffee_fortune()
elif fortune_type == 2:
    card_fortune()
else:
    print("Invalid choice.")
```

# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
with open("urunler.txt" , "a" , encoding= "utf-8") as file:
    ad = input("satın aldığınız ürün nedir?:")
    fiyat = input("satın aldığınız ürünün fiyatını giriniz:")
    file.write(f"{ad}\n {fiyat}\n")

print(f"{ad} ürünü {fiyat} TL fiyat bilgisiyle ürünler dosyasıne kaydedildi.")
````
